## Частина 5: Представлення даних

**Крок 1: Побудова графіку дата фрейму SF, використовуючи змінні X та Y**

```python
plt.plot(SF['X'],SF['Y'], 'ro')
plt.show()
```

Визначення номерів відділів поліції, складення словника pd_districts

```python
pd_districts = np.unique(SF['PdDistrict'])
pd_districts_levels = dict(zip(pd_districts, range(len(pd_districts))))
pd_districts_levels
```

Додавання цілого числа поліцейського відділу до нового стовпця
DataFrame

```python
SF['PdDistrictCode'] = SF['PdDistrict'].apply(lambda row: pd_districts_levels[row])
```

Використання створеного PdDistrictCode для автоматичної зміни кольору

```python
plt.scatter(SF['X'], SF['Y'], c=SF['PdDistrictCode'])
plt.show()
```

**Крок 2: Додання пакетів для побудови карт, щоб покращити сюжет**  <br/>
Обрання необхідних кольорів

```python
from matplotlib import colors
districts = np.unique(SF['PdDistrict'])
print(list(colors.cnames.values())[0:len(districts)])
```

Створення словника кольорів для кожного окружного відділу поліції.

```python
color_dict = dict(zip(districts, list(colors.cnames.values())[0:-1:len(districts)]
))
color_dict
```

Створення карти, використовуючи середні координати даних SF, щоб відцентрувати карту. Щоб зменшити час обчислення, використовується *plotEvery* для обмеження кількості побудованих даних.

```python
# Create map
map_osm = folium.Map(location=[SF['Y'].mean(), SF['X'].mean()], zoom_start = 12)
plotEvery = 50
obs = list(zip( SF['Y'], SF['X'], SF['PdDistrict']))
for el in obs[0:-1:plotEvery]:
    folium.CircleMarker(el[0:2], color=color_dict[el[2]], fill_color=el[2],radius=10).add_to(map_osm)
```
